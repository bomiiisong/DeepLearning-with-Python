# Time Series(시계열 데이터)

* KOSPI주가 데이터

## Ⅰ. KOSPI 주가 데이터 수집

> ### 1) Python Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

> ### 2) datetime Package

* 주가를 수집할 기간 지정

```python
from datetime import datetime

start = datetime(2010, 1, 1)
end = datetime(2020, 6, 30)
```

> ### 3) pandas_datareader Package

* Yahoo Finance에서 KOSPI 종목별 정보 수집
* 종목번호 : '035250.KS'

```python
import pandas_datareader as pdr  # 종목 정보 수집 패키지

GL = pdr.DataReader(name='035250.KS',
                    data_source = 'yahoo',
                    start=start,
                    end=end)
```

> ### 4) 수집된 주가정보 확인

* Close : 종가

```python
GL.head()
```

> ### 5) 종가('Close') 정보 시각화

```python
plt.figure(figsize = (14, 7))
plt.plot(GL['Close'])
plt.grid()
plt.show()
```

## Ⅱ. 종목 주가 예측 테스트

> ### 1) 예측 테스트를 위한 데이터셋

* 2010년 01월 01일 ~ 2019년 12월 31일

```python
GL_Trunc = GL[:'2019-12-31']
GL_Trunc.tail()
```

> ### 2) 날짜(GL_Trunc.index)와 종가('Close')로 데이터프레임생성

```python
DF = pd.DataFrame({'ds' : GL_Trunc.index, 'y' : GL_Trunc['Close']})

DF.reset_index(inplace=True)

del DF['Date']

DF.head()
```

> ### 3) fbprophet Package

* 페이스북에서 개발한 시계열 예측 패키지
* 경험적 규칙(Heuristic Rule)을 사용
    * Prophet 객체 생성 후 DF를 .fit()에 적용

```python
from fbprophet import Prophet

model = Prophet()
model.fit(DF)
```

* 예측 날짜 구간 생성
    * 2020년 07월 01일까지 생성
    * periods = 184

```python
X_test = model.make_future_dataframe(periods=184)
```

* 예측 날짜 구간 확인

```python
X_test.tail()
```

* 생성된 예측 날짜 구간으로 예측 실행

```python
pred = model.predict(X_test)
```

* 예측 결과 시각화
    * 산점도 : 원래값(y)
    * 실선 : 예측값(y_hat)

```python
model.plot(pred, figsize=(12, 6))
```

* 선형회귀 및 weekly, yearly 성분별 시각화

```python
model.plot_components(pred)
```

> ### 4) 실제값(Real)과 예측값(Forecast) 비교 시각화

```python
plt.figure(figsize = (14, 7))
plt.plot(GL.index, GL['Close'], label = 'Real')
plt.plot(pred['ds'], pred['yhat'], label = 'Forecast')
plt.grid()
plt.legend()
plt.show()
```
